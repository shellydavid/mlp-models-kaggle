```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

#################################################
'''

'''
def plot_curve(accuracy, loss):
    epochs = np.arange(loss.shape[0])
    plt.figure(figsize=(9,5)) 
    plt.subplot(1,2,1)
    plt.plot(epochs, accuracy)
    plt.xlabel("Epochs")
    plt.ylabel("Accuracy")
    plt.title("Training accuracy")

    plt.subplot(1,2,2)
    plt.plot(epochs, loss)
    plt.xlabel("Epochs")
    plt.ylabel("Loss")
    plt.title("Training loss")

    plt.show()

'''

'''
def plot_display_metrics(model, model_history):
    accuracy = np.array(model_history.history["accuracy"])
    loss = np.array(model_history.history["loss"])
    plot_curve(accuracy, loss)

    train_score = [accuracy[-1], loss[-1]]
    print(f"\nTraining accuracy: {round(train_score[0], 4) * 100}%")
    print(f"Training loss: {round(train_score[1], 3)}")



def plot_train_val(model_history):
    train_acc = np.array(model_history.history["accuracy"])
    train_loss = np.array(model_history.history["loss"])

    val_loss = np.array(model_history.history["val_loss"])
    val_acc = np.array(model_history.history["val_accuracy"])

    epochs = np.arange(train_loss.shape[0])
    plt.plot(epochs, train_loss, label = "Training")
    plt.plot(epochs, val_loss, label = "Validation")
    plt.xlabel("Epochs")
    plt.ylabel("Loss")
    plt.title("Training & Validation Loss")
    plt.legend()
    plt.show()

    train_score = [train_acc[-1], train_loss[-1]]
    val_score = [val_acc[-1], val_loss[-1]]

    print(f"\nTraining accuracy: {round(train_score[0], 4) * 100}%")
    print(f"Training loss: {round(train_score[1], 3)}")
    print(f"\nValidation accuracy: {round(val_score[0], 4) * 100}%")
    print(f"Validation loss: {round(val_score[1], 3)}")


def features_labels_split(data, label_column):
    features = data.copy()
    columns_titles = list(data.columns)
    features = features.reindex(columns=columns_titles)
    labels = features.pop(label_column)
    features = np.array(features)
    labels = np.array(labels)
    labels = labels.reshape((features.shape[0],1))
    return features, labels
```
